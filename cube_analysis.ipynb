```python
# show results of models processing by UMAP
from matplotlib import tight_layout
from matplotlib.pyplot import xlim
import umap

def show_results(results):
    for result in results:
        name = result[0]
        data = result[1].detach().cpu().numpy()
        print(name, data.shape)
        plt.figure(figsize=(20, 5), tight_layout=True)
        plt.title(name)
        for i in range(len(data)):
            processed = umap.UMAP().fit_transform(data[i])
            if i == 0:
                transform = umap.UMAP().fit(data[i])
            processed = transform.transform(data[i])
            if i == 0:
                xmin = processed[:, 0].min() * 1.1
                xmax = processed[:, 0].max() * 1.1
                ymin = processed[:, 1].min() * 1.1
                ymax = processed[:, 1].max() * 1.1
            # plt.subplot(1, len(data), i + 1, xlim=(xmin, xmax), ylim=(ymin, ymax))
            plt.subplot(1, len(data), i + 1)
            plt.scatter(processed[:, 0], processed[:, 1])

        # plt.colorbar()
        plt.show()
```

```python
import torch
import random
import data_loader
import cube_learn

_, test_loader = data_loader.StateLoader2(train_rate=0.5, batch_size=100, size=1000)

def gen_test_result(model_path, sample):
    result = torch.load(model_path, map_location='cpu')
    model = cube_learn.CubeGPTStateNet(n_layers=3)
    model.load_state_dict(result['model'])
    model.eval()

    # print(model)
    results = []
    for i in range(len(model.transformer_encoder.layers)):
        layer = model.transformer_encoder.layers[i]
        layer.self_attn.register_forward_hook(lambda self, input, output: results.append((f"self_attn{i}", output[0])))
        layer.norm1.register_forward_hook(lambda self, input, output: results.append((f"norm1{i}", output)))
        layer.linear1.register_forward_hook(lambda self, input, output: results.append((f"linear1{i}", output)))
        layer.norm2.register_forward_hook(lambda self, input, output: results.append((f"norm2{i}", output)))
        layer.linear2.register_forward_hook(lambda self, input, output: results.append((f"linear2{i}", output)))

    sample = torch.tensor(sample)
    sample = sample.permute(1, 0, 2)
    model(sample)
    return results

for sample, _ in test_loader:
    results = gen_test_result('runs/Sep13_09-10-02_5d14281e9a52/999.pth', sample)
    show_results(results)
    break
```
